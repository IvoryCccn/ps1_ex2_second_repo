```python
from data import load_gdp_data, clean_gdp_data, get_data_summary
from plot_utils import plot_gdp_trends, plot_log_gdp_trends, plot_growth_rates
```

```python
path = "../data/GDP.xlsx"

# Load data
gdp_df = load_gdp_data(path)
```

```python
# Data analysis
summary = get_data_summary(gdp_df)
print(f"Data shape: {summary['shape']}")
print(f"Columns: {summary['columns']}")
```

```python
# Clean data
cleaned_df = clean_gdp_data(gdp_df)
```

```python
# Data analysis
summary = get_data_summary(cleaned_df)
print(f"Data shape: {summary['shape']}")
print(f"Columns: {summary['columns']}")
```

```python
# Target countries
countries = ['Brazil', 'Switzerland', 'China', 'Germany', 'United Kingdom', 'United States', 'Japan']

# Generate plots
plot_gdp_trends(cleaned_df, countries)
plot_log_gdp_trends(cleaned_df, countries)
plot_growth_rates(cleaned_df, countries)
```

```python
# Calculate average growth rates
print("\nAverage GDP Growth Rates (2000-2022):")
for country in countries:
    if country in cleaned_df.columns:
        growth_rates = cleaned_df[country].pct_change().dropna() * 100
        avg_growth = growth_rates.mean()
        print(f"{country}: {avg_growth:.2f}%")
```
